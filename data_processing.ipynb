# Normalizing the data


```python
import pandas as pd
```

Importing all the datasets. 

```python
listings_1 = pd.read_csv('data/listings_1.csv')
```

```python
listings_2 = pd.read_csv('data/listings_2.csv')
```

```python
listings_3 = pd.read_csv('data/listings_3.csv')
```

```python
listings_4 = pd.read_csv('data/listings_4.csv')
```

Concatenating the data into one single dataset. 

```python
listings = pd.concat([listings_1, listings_2, listings_3, listings_4], axis=0, ignore_index=True)
```

```python
listings.head()
```

```python
listings.info()
```

Looking for null values. 

```python
listings.isnull().sum()
```

Finding and dropping the columns that have missing values. 

```python
ls = [col for col in listings.columns if listings[col].isnull().sum() > 1500 and not col in ['bedrooms', 'bathrooms']]
ls
```

```python
listings.drop(['host_location'
               , 'host_about'
               , 'neighbourhood'
               , 'description'
               , 'neighborhood_overview'
               , 'neighbourhood_group_cleansed'
               , 'calendar_updated'
               , 'license'],axis=1, inplace=True)
```

```python
listings.head()
```

Looking at the reviews data. 

```python
for col in listings.columns:
    if 'review' in col:
        print(listings[col].value_counts())
```

Export the clean data. 

```python
listings.to_csv("listing_clean.csv")
```

Extracting host columns into separate table. 

```python
host = listings[['host_id'
        , 'host_name'
        , 'host_since'
        , 'host_is_superhost'
        , 'host_response_time'
        , 'host_response_rate'
        , 'host_acceptance_rate'
        , 'host_listings_count'
        , 'host_total_listings_count'
        , 'host_identity_verified']]
```

Dropping duplicates from the host table.

```python
host.drop_duplicates(inplace=True)
```

```python
host.shape
```

Extracting review columns in to separate table.

```python
review = listings[['id',
                   'review_scores_rating'
                  ,'review_scores_accuracy'
                  ,'review_scores_cleanliness'
                  ,'review_scores_checkin'
                  ,'review_scores_communication'
                  ,'review_scores_location'
                  ,'review_scores_value']]
```

```python
review.shape
```

Extracting review coumns into a separte table. 

```python
geo_location = listings[['id'
         , 'latitude'
         , 'longitude'
         , 'neighbourhood_cleansed']]
```

```python
geo_location.shape
```

```python
geo_location.isnull().sum()
```

Extracting neighbourhood coumns into a separte table. 

```python
neighbourhood = listings['neighbourhood_cleansed'].unique()
```

```python
sorted(neighbourhood)
```

Creating index for the neighbourhood. 

```python
neighbourhoods = pd.DataFrame({'n_id': [i+1 for i in range(len(neighbourhood))]
                             ,'neighbourhood': sorted(neighbourhood)})
neighbourhoods
```

Creating a copy of the dataset to be safe. 

```python
safe_copy = listings.copy(deep=True)
```

```python
listings.drop(['latitude'
         , 'longitude'], axis=1, inplace=True)
```

```python
listings.head()
```

```python
safe_copy.head()
```

```python
listings.info()
```

```python
review['reviews_per_month'] = listings['reviews_per_month']
```

```python
review.head()
```

```python
review.dropna(inplace=True)
```

```python
review.to_csv("air_review.csv")
```

```python
listings.drop(['reviews_per_month'], axis=1, inplace=True)
```

```python
listings.info()
```

```python
listings.drop([col for col in listings.columns if 'calculated' in col], axis=1, inplace=True)
```

```python
listings.drop([col for col in listings.columns if 'host' in col and not col == 'host_id'], axis=1, inplace=True)
```

```python
listings.drop([col for col in listings.columns if 'review' in col], axis=1, inplace=True)
```

```python
listings.info()
```

```python
room_typ = sorted(list(set(listings['room_type'])))
room_type = pd.DataFrame({'room_type_id': [i+1 for i in range(len(room_typ))]
                         ,'room_type': room_typ})

room_type
```

```python
property_typ = sorted(list(set(listings['property_type'])))
property_type = pd.DataFrame({'property_type_id': [i+1 for i in range(len(property_typ))]
                         ,'property_type': property_typ})

property_type
```

```python
listings.drop(['listing_url'
               , 'scrape_id'
               , 'last_scraped'
               , 'source'
               , 'picture_url'
               , 'amenities'
               , 'calendar_last_scraped'
               , 'bathrooms_text'
               , 'beds'
               , 'neighbourhood_cleansed'], axis=1, inplace=True)
```

```python
listings.info()
```

```python
tables = {'listing': listings
          ,'host': host
          ,'review': review
          ,'geo_location': geo_location
          ,'room_type': room_type
          ,'property_type': property_type
          ,'neighbourhood': neighbourhoods}

for table in tables.keys():
    tables[table].to_csv(f'air_{table}.csv', index=False)
```
